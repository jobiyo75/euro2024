```python
def convert_jpg_to_csv(path):
    if path != None:
        data = pd.read_csv(path)
        df = pd.DataFrame(data)
        # delete columns
        df.drop(columns=['page'], inplace=True)
        df.drop(columns=['original_filename'], inplace=True)
        df.drop(columns=['file_url'], inplace=True)
        df.drop(columns=['platform_link'], inplace=True)

        return df
    else:
        return 'path not exist'
```
